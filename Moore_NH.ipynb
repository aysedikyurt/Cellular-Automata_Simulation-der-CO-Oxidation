```python
from IPython.core.display import clear_output
#a:infected cells
#b:sicked cells
#Moore Neighborhood
clear_output
n=100 #maximum Sickdegree(ill)
k1=1
k2=2
g=15 #speed of infection

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

length=50
size=(length,length)



#A=np.full((length,length),0)
#A[10,10]=30
#np.random.seed(42)
A = np.random.randint(0, high=n+1, size=size, dtype=int)



print(A)
plt.gca().invert_yaxis()
#cmap = colors.ListedColormap(['lightgreen','pink','red','blue','purple','black'])

plt.pcolor(A, cmap='rainbow', vmin=0, vmax=n)
plt.colorbar()
plt.show()
```

```python
from numpy import infty
def newhob(A):
  B=np.copy(A)
  for r in range(0,length):
    for j in range(0,length):
      #a:infected cell
      #b:sicked cell
      #s:sum of degree
      a=0
      b=0
      s=0

      if r==0:
        if j==0:
          #A[0,0] köşesi
          if 0<A[r,j+1]<n:
            a=a+1
            s=s+A[r,j+1]
          b += A[r,j+1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r+1,j]<n:
            a=a+1
            s=s+A[r+1,j]
          b += A[r+1,j]==n
          if 0<A[r+1,j+1]<n:
            a=a+1
            s=s+A[r+1,j+1]
          b += A[r+1,j+1]==n




        elif j==length-1:
          #A[0,9] köşesi
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r+1,j-1]<n:
            a=a+1
            s=s+A[r+1,j-1]
          b += A[r+1,j-1]==n
          if 0<A[r+1,j]<n:
            a=a+1
            s=s+A[r+1,j]
          b += A[r+1,j]==n
        else:
          #r=0 kenarı ama köşeler yok.
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r+1,j-1]<n:
            a=a+1
            s=s+A[r+1,j-1]
          b += A[r+1,j-1]==n
          if 0<A[r+1,j]<n:
            a=a+1
            s=s+A[r+1,j]
          b += A[r+1,j]==n
          if 0<A[r,j+1]<n:
            a=a+1
            s=s+A[r,j+1]
          b += A[r,j+1]==n
          if 0<A[r+1,j+1]<n:
            a=a+1
            s=s+A[r+1,j+1]
          b += A[r+1,j+1]==n

      elif r==length-1:
        if j==0:
          #A[9,0] köşesi
          if 0<A[r-1,j]<n:
            a=a+1
            s=s+A[r-1,j]
          b += A[r-1,j]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r-1,j+1]<n:
            a=a+1
            s=s+A[r-1,j+1]
          b += A[r-1,j+1]==n
          if 0<A[r,j+1]<n:
            a=a+1
            s=s+A[r,j+1]
          b += A[r,j+1]==n

        elif j==length-1:
          #A[9,9] köşesi
          if 0<A[r-1,j-1]<n:
            a=a+1
            s=s+A[r-1,j-1]
          b += A[r-1,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r-1,j]<n:
            a=a+1
            s=s+A[r-1,j]
          b += A[r-1,j]==n
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
        else:
          #r=9 kenarı köşeler hariç
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r,j+1]<n:
            a=a+1
            s=s+A[r,j+1]
          b += A[r,j+1]==n
          if 0<A[r-1,j+1]<n:
            a=a+1
            s=s+A[r-1,j+1]
          b += A[r-1,j+1]==n
          if 0<A[r-1,j]<n:
            a=a+1
            s=s+A[r-1,j]
          b += A[r-1,j]==n
          if 0<A[r-1,j-1]<n:
            a=a+1
            s=s+A[r-1,j-1]
          b += A[r-1,j-1]==n


      elif j==0:
        #j=0 kenarı köşeler hariç
        if 0<A[r-1,j]<n:
          a=a+1
          s=s+A[r-1,j]
        b += A[r-1,j]==n
        if 0<A[r,j]<n:
          a=a+1
          s=s+A[r,j]
        b += A[r,j]==n
        if 0<A[r,j+1]<n:
          a=a+1
          s=s+A[r,j+1]
        b += A[r,j+1]==n
        if 0<A[r-1,j+1]<n:
          a=a+1
          s=s+A[r-1,j+1]
        b += A[r-1,j+1]==n
        if 0<A[r+1,j]<n:
          a=a+1
          s=s+A[r+1,j]
        b += A[r+1,j]==n
        if 0<A[r+1,j+1]<n:
          a=a+1
          s=s+A[r+1,j+1]
        b += A[r+1,j+1]==n


      elif j==length-1:
          #j=9 kenarı köşeler hariç.
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r+1,j-1]<n:
            a=a+1
            s=s+A[r+1,j-1]
          b += A[r+1,j-1]==n
          if 0<A[r+1,j]<n:
            a=a+1
            s=s+A[r+1,j]
          b += A[r+1,j]==n
          if 0<A[r-1,j]<n:
            a=a+1
            s=s+A[r-1,j]
          b += A[r-1,j]==n
          if 0<A[r-1,j-1]<n:
            a=a+1
            s=s+A[r-1,j-1]
          b += A[r-1,j-1]==n

      else:
         #ortadakiler
          if 0<A[r,j-1]<n:
            a=a+1
            s=s+A[r,j-1]
          b += A[r,j-1]==n
          if 0<A[r,j]<n:
            a=a+1
            s=s+A[r,j]
          b += A[r,j]==n
          if 0<A[r+1,j-1]<n:
            a=a+1
            s=s+A[r+1,j-1]
          b += A[r+1,j-1]==n
          if 0<A[r+1,j]<n:
            a=a+1
            s=s+A[r+1,j]
          b += A[r+1,j]==n
          if 0<A[r-1,j]<n:
            a=a+1
            s=s+A[r-1,j]
          b += A[r-1,j]==n
          if 0<A[r-1,j-1]<n:
            a=a+1
            s=s+A[r-1,j-1]
          b += A[r-1,j-1]==n
          if 0<A[r+1,j+1]<n:
            a=a+1
            s=s+A[r+1,j+1]
          b += A[r+1,j+1]==n
          if 0<A[r,j+1]<n:
            a=a+1
            s=s+A[r,j+1]
          b += A[r,j+1]==n
          if 0<A[r-1,j+1]<n:
            a=a+1
            s=s+A[r-1,j+1]
          b += A[r-1,j+1]==n





        #hücre durumları
      if A[r,j]==0:
          B[r,j]= int((a/k1)+(b/k2))
      elif A[r,j]==n:
          B[r,j]=0
      else:
          B[r,j]=int(s/a)+g
          if B[r,j]>n:
            B[r,j]=n
  return B

```
