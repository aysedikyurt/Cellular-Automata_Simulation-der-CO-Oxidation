```python
#a:infected cells
#b:sicked cells
#Neumann Neighborhood
n=100 #maximum Sickdegree(ill)
k1=1
k2=2
g=15 #speed of infection

import matplotlib.pyplot as plt
#import seaborn as sns
import numpy as np
from matplotlib import colors

length=50
size=(length,length)



#A=np.full((length,length),0)
#A[10,10]=30
#np.random.seed(42)
A = np.random.randint(0, high=n+1, size=size, dtype=int)
#np.random.choice(A, 5, p=[0.5, 0.2, 0.3])
#A=np.array(range(101)).reshape(20,20)
#A = np.random.randn(0, high=n+1, size=(size))
#A = np.random.uniform(low=0, high=1, size=(size))


print(A)
#sns.heatmap(A)
plt.gca().invert_yaxis()
#cmap = colors.ListedColormap(['lightgreen','pink','red','blue','purple','black'])

plt.pcolor(A, cmap='rainbow', vmin=0, vmax=n)
#plt.clim(0, length)
plt.colorbar()
plt.show()
#print(A[0,3])
#x=A[number()>=37]
#A.sort()
#count=size(size)-bisect(A,2)
#print(count)
```

```python
#from typing import Counter

def hopostep(A):
  B=np.copy(A)
  for r in range(0, length):
    for j in range(0, length):
      #Şu an için kenarlar katılmadı. Onların bir ya da iki komşusu olcak.

    #  a=len([1 for i in A if 0 > i > n-1])

    #a:infected cell (etrafındaki hücreleri syaamıyorum)
      a=0
      if not r==length-1:
        if 0<A[r+1,j]<n:
            a=a+1
      if not r==0:
        if  0<A[r-1,j]<n:
            a=a+1
      if not j==length-1:
        if 0<A[r,j+1]<n:
            a=a+1
      if not j==0:
        if 0<A[r,j-1]<n:
            a=a+1
      if 0<A[r,j]<n:
            a=a+1

      #print(a)

      #b=sicked cell
      b=0
      if not r==length-1:
        if A[r+1,j]==n:
            b=b+1
      if not r==0:
        if  A[r-1,j]==n:
            b=b+1
      if not j==length-1:
        if A[r,j+1]==n:
            b=b+1
      if not j==0:
        if A[r,j-1]==n:
            b=b+1
      if 0<A[r,j]<n:
          b=b+1


      #print(b)

      #a=A.allcount(0<A[r-1,j]<n)+A.allcount(0<A[r+1,j]+A.allcount(0<A[r,j-1]<n)+A.allcount(0<A[r,j+1]<n)<n) #infected cells number

      #b=count(A[r-1,j]==n and A[r+1,j]==n and A[r,j-1]==n and A[r,j+1]==n)  #ill cells number


      s=A[r,j]
      if not r==length-1:
        if not A[r+1,j]==100:
          s=s+A[r+1,j]

      if not r==0:
        if not A[r-1,j]==100:
          s=s+A[r-1,j]

      if not j==length-1:
        if not A[r,j+1]==100:
          s=s+A[r,j+1]

      if not j==0:
        if not A[r,j-1]==100:
          s=s+A[r,j-1]


      #s=A[r,j]+A[r+1,j]+A[r-1,j]+A[r,j-1]+A[r,j+1] #sum of the state number

      if A[r,j]==0:
          B[r,j]= int((a/k1)+(b/k2))
      elif A[r,j]==n:
          B[r,j]=0
      else:
          B[r,j]= int(s/a)+g
          if B[r,j]>n:
            B[r,j]=n




          #n den büyükse bu sayı n olmalı.
  return B
          #a saymalı
          #skenarlar için tekrar düenlemeli
         # mod bölmeden kalnala yapabilrsin kenarları belki
        #  uzunluğu düenle

```

```python
from PIL import Image
import glob
import time

# create an empty list called images
images = []

# get the current time to use in the filename
timestr = time.strftime("%Y%m%d-%H%M%S")

# get all the images in the 'images for gif' folder
for filename in sorted(glob.glob('norm/*.png')): # loop through all png files in the folder
    im = Image.open(filename) # open the image
    im_small = im.resize((1200, 1500), resample=0) # resize them to make them a bit smaller
    images.append(im_small) # add the image to the list

# calculate the frame number of the last frame (ie the number of images)
last_frame = (len(images))

# create 10 extra copies of the last frame (to make the gif spend longer on the most recent data)
for x in range(0, 9):
    im = images[last_frame-1]
    images.append(im)

# save as a gif
images[0].save('normgif/norm' + timestr + '.gif',
               save_all=True, append_images=images[1:], optimize=False, duration=500, loop=0)
```
